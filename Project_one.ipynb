```python
import pandas as pd
link = "https://en.wikipedia.org/wiki/List_of_African_countries_by_population"
table = pd.read_html(link)
len(table)
```

```python
df = table[0]
df.head()
```

```python
def percent(text):
    text = text.replace("%","")
    flt = float(text)
    return flt/100

df['% Africa'] = df['% Africa'].apply(percent)
df['% growth'] = df['% growth'].apply(percent)
print(df.info())
```

```python
df.tail()
```

```python
import re
string = "21 Mar 2022[5]"
def clean_dates(string):
    string = str(string)
    return re.sub("\[[0-9]{,2}\]","",string)
df['clean_date'] = df['Official date'].apply(clean_dates)
df.head(10)
```

```python
df['clean_date'] = pd.DatetimeIndex(df['clean_date'])
df = df.drop(['Unnamed: 0','Unnamed: 7','Official date'],axis = 1)
df.head()
```
