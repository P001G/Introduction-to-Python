```python
from PyPDF2 import PdfReader, PdfWriter
def unlock_pdf(pdf_path, password):
    # Create a PDF reader object
    pdf_reader = PdfReader(pdf_path)
    
    # Check if the PDF is encrypted
    if pdf_reader.is_encrypted:
        # Try to decrypt the PDF
        pdf_reader.decrypt(password)
        
        # Create a PDF writer object
        pdf_writer = PdfWriter()
        
        # Add all pages to the writer
        for page_num in range(len(pdf_reader.pages)):
            pdf_writer.add_page(pdf_reader.pages[page_num])
            
        # Write the decrypted content to a new PDF file
        with open('unlocked.pdf', 'wb') as new_pdf_file:
            pdf_writer.write(new_pdf_file)
        return 'unlocked.pdf'
    else:
        # PDF is not encrypted, no need to unlock
        return pdf_path

    
def pdf_to_excel(pdf_path, output_dir):
    # Use tabula to extract tables from PDF
    tables = tabula.read_pdf(pdf_path, pages='all', multiple_tables=True)
    all = []
    # Convert each table to Excel
    for i, table in enumerate(tables):
        print(f"----------------> read table {i}")
        excel_file = os.path.join(output_dir, f'table_{i+1}.xlsx')
        table.to_excel(excel_file, index=False)
        if table.shape[0]>5:
            all.append(table)
        print(f"Table {i+1} converted to Excel: {excel_file}")
    final = pd.concat(all)
    final.to_excel("FinalDataset.xlsx")
```
