# <font color = "gold"> Merging,Concatenation </font>

```python
import pandas as pd
y1 = pd.DataFrame({
    "id":[1,2,3,4,5],
    'name':['john','william','alice','alex','james']
    
})

y2 = pd.DataFrame({
    "id":[1,4,5],
    'Age':[23,67,34]
    
})

display(y1)
display(y2)
```

```python
y1.merge(y2,on = 'id',how = 'left')
```

```python
y1.merge(y2,on = 'id',how = 'right')
```

```python
y1.merge(y2,on = 'id',how = 'inner')
```

```python

y3 = pd.DataFrame({
    "id":[6,7],
    'name':['agnes','maryanne']
    
})
display(y1)
display(y3)


pd.concat((y1,y3),axis =0)
```

# melting data

```python

weekly.melt(id_vars = 'names') # long format
```
