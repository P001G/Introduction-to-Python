```python
def Counter(items):
    total = {}
    
    for item in items:
        print(total)
        if item in total:
            total[item]+=1
        else:
            total[item] =1
    return total
items = ['John',"John",'Paul']
from collections import Counter

def mode(items):
    y = Counter(items).most_common()
    return y[0][0]

mode(['john','john','john','paul','paul','alex'])
```

```python
# grou by  -  pivoting
# we waht to know what was the average age of those that died
df.groupby(['Survived'])['Fare'].agg(['mean',"median","count",mode]) # the groupby in python are more robust
# which gender survived more
df.groupby(['Sex','Survived'])["Id"].agg(['count'])
# whicg pclass survived more
df.groupby(['Pclass','Survived'])["Id"].agg(['count'])
# for each gender that survived or died what what was the average fare they paid
df.pivot_table(index = 'Sex',columns = 'Survived',values = 'Fare',aggfunc = 'mean')
```

```python
# using for loops
# Method one
survived  = [] 
for i in df['Survived']:
    if i==0:
        survived.append('died')
    else:
        survived.append('lived')
df['Survival'] = survived
print(df['Survival'].value_counts())
df.head()
```

```python
# list comprehension
# method  -  1.5  - list comprehension
df['list_survived'] = ['dead' if i==0 else 'alive' for i in df['Survived']]
df.head()
```

```python
# method 2
df['copy'] = df['Survived']
df['copy'] =df['copy'].replace({1:'alive',0:'dead'})
df.head()
```

```python
# method 3 write a function

def survive(x):
    if x==1:
        return 'alive'
    else:
        return 'dead'

df['func_survived'] = df['Survived'].apply(survive) # one way of the using the function
df['func_survived1'] = list(map(survive,df['Survived'])) # map apply (uses parallel computing so better for big data)
df.head()
```
